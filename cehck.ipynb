```python
df = indicator.Baseline(SSL_df,period=8)
df.head(10)
```

```python
SSL_df = df.iloc[100:200].head(60)
```

```python
import plotly.graph_objects as go

# Add candlestick trace
fig = go.Figure(go.Candlestick(
    x=SSL_df['Time'],
    open=SSL_df['Open'],
    high=SSL_df['High'],
    low=SSL_df['Low'],
    close=SSL_df['Close'],
    line=dict(width=1),
    opacity=1,
    increasing_fillcolor='#24A06B',
    decreasing_fillcolor="#CC2E3C",
    increasing_line_color='#2EC886',
    decreasing_line_color='#FF3A4C'
))

# Add SSL lines
fig.add_trace(go.Scatter(
    x=SSL_df['Time'], 
    y=SSL_df['sslDown'], 
    name="SSLDOWN", 
    line=dict(color='green')
))
fig.add_trace(go.Scatter(
    x=SSL_df['Time'], 
    y=SSL_df['sslUp'], 
    name="SSLUP", 
    line=dict(color='blue')
))

# Filter and add annotations for Buy signals
buy_signals = SSL_df[SSL_df['SIGNAL'] == 'BUY']
fig.add_trace(go.Scatter(
    x=buy_signals['Time'], 
    y=buy_signals['sslDown'],
    mode='markers',
    marker=dict(symbol='triangle-up', size=10, color='green'),
    name='BUY Signal'
))

# Filter and add annotations for Sell signals
sell_signals = SSL_df[SSL_df['SIGNAL'] == 'SELL']
fig.add_trace(go.Scatter(
    x=sell_signals['Time'], 
    y=sell_signals['sslUp'],
    mode='markers',
    marker=dict(symbol='triangle-down', size=10, color='red'),
    name='SELL Signal'
))

# Update the layout of the figure
fig.update_layout(
    width=1000, height=400,
    margin=dict(l=10, r=10, b=10, t=10),
    font=dict(size=10, color="#e1e1e1"),
    paper_bgcolor="#1e1e1e",
    plot_bgcolor="#1e1e1e",
    xaxis_rangeslider_visible=False
)

# Update axes properties
fig.update_xaxes(gridcolor="#1f292f", showgrid=True, fixedrange=True)
fig.update_yaxes(gridcolor="#1f292f", showgrid=True)

# Display the figure
fig.show()
```
